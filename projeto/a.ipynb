```python
import altair as alt
import pandas as pd
import numpy as np
import scipy as scp
alt.data_transformers.disable_max_rows()
```

```python
data = pd.read_csv("teste.csv")
xdata = np.arange(0,3735)
data["a"] = xdata
data
```

```python
alt.Chart(data).mark_point(color = "blueviolet").encode(
    alt.X("a"),
    alt.Y("erro"),
    tooltip = ["erro"]
).properties(
    width = 2000
)
```

```python
df = pd.read_csv("ppp.csv")
df
df = df.loc[df["temperatura atual"]!= 1.31]
```

```python
alt.Chart(df).mark_point().encode(
    alt.X(" tempo"),
    alt.Y("temperatura atual", scale = alt.Scale(domain = [295,337])),
    tooltip = ["temperatura atual"]
).properties(
    width = 1000
)
```
