```python
import pandas as pd
import ajuste_sen as aj
import numpy as np
from scipy.optimize import curve_fit
```

```python
def process(path,argw=[1.5, 3.14, 0, 2],channel_ys = "CH2V"):
    if channel_ys == "CH1V":
        color_point = "green"
    else:
        color_point = "blueviolet"
    df = pd.read_csv(path)

    df["Time(s)"]*=1000

    r_data,param,cov = aj.regres(df,arg = argw,channel_y=channel_ys)

    a1,a2 = aj.data_x_regression(df,r_data,channel_y=channel_ys,color_p=color_point)

    return a1, a2, param

path = "teste.csv"
a1,a2,param = process(path,channel_ys="CH1V")
a1.display()
(a1+a2).display()

a1,a2,param = process(path)
a1.display()
(a1+a2).display()

Amplitude, freq, fase, Voff = param
```

```python
with open("test.txt", "w") as f:
    f.write(f"Amplitude = {Amplitude}\n")
    f.write(f"Frequencia = {freq}\n")
    f.write(f"Fase = {fase}\n")
    f.write(f"Voff = {Voff}\n")
```

```python
path = "RC10.csv"
def save_datas(path_data,path_save,argww = [3,3.14,-1,10]):
    a1,a2,parametros_Vc = process(path_data,argw = [3, 3.14, -1, 10])
    avc,bvc,cvc,dvc = parametros_Vc
    a1.display()
    (a1+a2).display()

    a1,a2,parametros_Vg = process(path_data,argw = [3, 3.14, -100, -10],channel_ys="CH1V")
    avg,bvg,cvg,dvg = parametros_Vg
    a1.display()
    (a1+a2).display()

    with open(path_save,"w") as f:
        f.write(f"funcao Vg = {avg}*sen({bvg}x + {cvg}) + {dvg}\n")
        f.write(f"funcao Vc = {avc}*sen({bvc}x + {cvc} + {dvc})\n")
        f.write(f"Amplitude Vr {avc-avg}\n")
        f.write(f"Fase Vr {cvc - cvg}")

save_datas(path,"z RC - 740.txt")
```
